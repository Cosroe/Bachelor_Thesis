```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from adjustText import adjust_text
```

```python
# Setting the desired Seaborn theme
sns.set_theme(
    context="paper",
    style='ticks',
    palette='colorblind',
    font='monospace',
    font_scale=1.2,
    rc={
        "figure.figsize": (12, 8),
        "axes.titlesize": "large",
        "axes.labelsize": "medium",
        "xtick.labelsize": "small",
        "ytick.labelsize": "small",
        "legend.fontsize": "medium",
    }
    
    )
```

# $\beta$ delta_EF vs. EF Plots

```python
mydata = pd.read_csv("MG_diff_coefficients_and_EF.csv")
mydata
```

```python
# Convert columns to numeric, forcing errors to NaN to handle mixed types
mydata['RL'] = pd.to_numeric(mydata['RL'], errors='coerce')
mydata['GS'] = pd.to_numeric(mydata['GS'], errors='coerce')
mydata['RE'] = pd.to_numeric(mydata['RE'], errors='coerce')
mydata['MO'] = pd.to_numeric(mydata['MO'], errors='coerce')
```

```python
mydata.info()
```

```python
# Function to create the plot and add text annotations
def create_plot(ax, x, y, xlabel, ylabel, title):
    sns.regplot(x=x, y=y, data=mydata, ax=ax, color='lightgray', marker='.', lowess=True)
    sns.regplot(x=x, y=y, data=mydata, ax=ax, color='black', marker='.' , ci=False , line_kws={'alpha':0.3})
    ax.set_title(title)
    ax.set_xlabel(xlabel)
    ax.set_ylabel(ylabel)
    texts = [ax.text(row[x] + 0.2, row[y], row['Abbreviation'], horizontalalignment='center', verticalalignment='bottom', size='small', color='black', weight='normal') 
             for idx, row in mydata.iterrows()]
    adjust_text(texts, ax=ax, arrowprops=dict(arrowstyle="->", color='black', lw=0.5))

# Plot RL vs delta_RL
plt.figure(figsize=(10, 8))
create_plot(plt.gca(), 'RL', 'delta_RL', r'$RL$', r'$\beta_{3,i}$', r'$RL$ vs $\beta_{3,i}$')
plt.show()

# Plot GS vs delta_GS
plt.figure(figsize=(10, 8))
create_plot(plt.gca(), 'GS', 'delta_GS', r'$GS$', r'$\beta_{4,i}$', r'$GS$ vs $\beta_{4,i}$')
plt.show()

# Plot RE vs delta_RE
plt.figure(figsize=(10, 8))
create_plot(plt.gca(), 'RE', 'delta_RE', r'$RE$', r'$\beta_{5,i}$', r'$RE$ vs $\beta_{5,i}$')
plt.show()

# Plot MO vs delta_MO
plt.figure(figsize=(10, 8))
create_plot(plt.gca(), 'MO', 'delta_MO', r'$MO$', r'$\beta_{6,i}$', r'$MO$ vs $\beta_{6,i}$')
plt.show()
```
